# Средний чек в ресторанах по РФ, Москве и Санкт-Петербургу

```python
import pandas as pd
```

```python
data_russia = pd.read_csv("amCharts.csv")  
df_r = data_russia.rename(columns={"Россия": "average"})
```

```python
data_moscow= pd.read_csv("amCharts_Moscow.csv")
df_m = data_moscow.rename(columns={"Москва": "average"})
```

```python
df_m.info()
```

```python
data_peter= pd.read_csv("amCharts_Peter.csv")
df_p = data_peter.rename(columns={"Санкт-Петербург": "average"})
```

```python
#Объединяем данные по Питеру, Москве, России
merged_df = pd.concat([df_r, df_m])
df = pd.concat([merged_df, df_p])
```

```python
df.info()
```

```python
#Форматируем столбец даты в формат даты 
df['date'] = df.date.astype('datetime64[ns]')
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
#Сравним данные по среднему чеку по Москве, Санкт-Петербургу и России

df_mos =df.loc[df['region'] == 'Москва'] 
df_ru =df.loc[df['region'] == "Россия"]
df_pi =df.loc[df['region'] == "Санкт-Петербург"]

trace1 = go.Scatter(x = df_mos.date, y = df_mos.average, mode = 'lines', name = 'Москва')
trace2 = go.Scatter(x = df_ru.date, y = df_ru.average, mode = 'lines', name = 'Россия')
trace3 = go.Scatter(x = df_pi.date, y = df_pi.average, mode = 'lines', name = 'Санкт-Петербург')
data = [trace1, trace2, trace3]

layout = dict(title = "Средний чек в ресторанах")
fig = dict(data = data, layout = layout)
iplot(fig, show_link=False)
```
